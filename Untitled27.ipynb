Q9.  What is Tableau Prep, and why is it important in real-world data projects?

What is Tableau Prep, and why is it important in real-world data projects?

Then explain how you would:

Remove null values

Remove duplicate records

Change incorrect data types (Answer in steps, no tool access required.)

Sol :

Tableau Prep is a data preparation tool used to clean, transform, and organize data before analysis.

Importance:

- Improves data quality.
- Saves analysis time.
- Ensures accurate reporting.

Remove Null Values

i . Open Tableau Prep.

ii . Identify fields containing Null values.

iii . Select Null records.

iv . Filter or remove them.

 Remove Duplicate Records

i . Review data using Profile Pane.

ii . Identify duplicate rows.

iii . Use Aggregate or Clean Step.

iv . Keep unique records only.

Change Incorrect Data Types

i . Select the field.

ii . Click the data type icon.

iii . Choose the correct type (String, Number, Date, etc.).

iv . Verify the results.


Q10.  A retail manager wants answers to the following:

Which category generates the highest profit?

Which region performs poorly?

How does sales change over time?

a) Mention which charts you would use
b) Identify one key insight that management could derive
c) Explain how this helps in business decision-making

Sol :

a) Charts to use: Use a Bar Chart for Category vs. Profit, a Map for Region performance, and a Line Chart for Sales changes over time.

b) Key insight: Management could discover that while a specific category drives high sales volume, it is actually generating negative profit in a particular region.

c) Business decision-making: These insights allow the manager to reallocate resources effectively, such as cutting marketing spend for unprofitable product lines or investigating operational inefficiencies in poorly performing regions.
